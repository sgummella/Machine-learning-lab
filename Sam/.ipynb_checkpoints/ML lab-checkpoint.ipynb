#### Problem 1: Dataset Import & Cleaning
Check **"Profit"** and **"Sales"** in the dataset, convert these two columns to numeric type. 


```python
orders['Profit'] = orders['Profit'].str.replace(',', '')
orders['Profit'] = orders['Profit'].str.replace('$', '')

orders['Sales'] = orders['Sales'].str.replace(',', '')
orders['Sales']= orders['Sales'].str.replace('$', '')

orders['Profit'] = pd.to_numeric(orders['Profit'])
orders['Sales'] = pd.to_numeric(orders['Sales'])
```

#### Problem 2: Inventory Management
- Retailers that depend on seasonal shoppers have a particularly challenging job when it comes to inventory management. Your manager is making plans for next year's inventory.
- He wants you to answer the following questions:
    1. Is there any seasonal trend of inventory in the company?
    2. Is the seasonal trend the same for different categories?

```python
orders['Month'] = orders['Order.Date'].apply(lambda x: int(x.split('/')[0]))
```

```python
by_month_cat = orders.groupby(['Month', 'Category']).agg({'Quantity' : 'sum'})
```

```python
by_month_cat.sort_values('Quantity', ascending = False)
```

```python
plt.plot(by_month_cat.unstack())
```

Christmas time (November, December) clearly the most popular for each different category, interestingly Jan/Feb then the least orders coming in. Overall orders increasing (hence inventory decreasing) throughout the year


### Problem 3: Why did customers make returns?
- Your manager required you to give a brief report (**Plots + Interpretations**) on returned orders.

	1. How much profit did we lose due to returns each year?

	2. How many customer returned more than once? more than 5 times?

	3. Which regions are more likely to return orders?

	4. Which categories (sub-categories) of products are more likely to be returned?

```python
returns= pd.read_csv("data/Returns.csv")
```

```python
print(returns.shape)
returns.head()
```

```python
returns_merged = pd.merge(returns, orders, how = 'inner', left_on = 'Order ID', right_on = 'Order.ID')
```

```python
returns_merged['Year'] = returns_merged['Order.Date'].apply(lambda x: int(x.split('/')[2]))
```

```python
profit_lost = returns_merged.groupby(['Year']).agg({'Profit' : 'sum'})
```

```python
plt.plot(profit_lost)
```

```python
customer_returns = returns_merged.groupby(['Customer.ID']).agg({'Order.ID':'count'})
```

```python
customer_returns[customer_returns['Order.ID'] > 1].sort_values(by= 'Order.ID', ascending = False).shape[0]
```

```python
customer_returns[customer_returns['Order.ID'] > 5].sort_values(by= 'Order.ID', ascending = False).shape[0]
```

```python
returns_merged
```

```python
region_returns = returns_merged.groupby(['Region_y']).agg({'Order.ID':'count'}).rename({'Order.ID': 'total_returns'})
```

```python
region_orders = orders.groupby(['Region']).agg({'Order.ID':'count'}).rename({'Order.ID': 'total_orders'})
```

```python
orders_and_returns = pd.concat([region_orders, region_returns], axis = 1) 
```

```python
pd.DataFrame((orders_and_returns.iloc[:,1] / orders_and_returns.iloc[:,0]) * 100).sort_values(by = 'Order.ID', ascending = False).rename(columns = {'Order.ID': 'Returns_prop'})
```

```python
subcat_returns = returns_merged.groupby(['Sub.Category']).agg({'Order.ID':'count'})
```

```python
subcat_orders = orders.groupby(['Sub.Category']).agg({'Order.ID':'count'})
```

```python
o_and_r_subcat = pd.concat([subcat_orders, subcat_returns], axis = 1) 
```

```python
pd.DataFrame((o_and_r_subcat.iloc[:,1] / o_and_r_subcat.iloc[:,0]) * 100).sort_values(by = 'Order.ID', ascending = False).rename(columns = {'Order.ID': 'Returns_prop'})
```
