### Problem 1: Dataset Import & Cleaning
Check **"Profit"** and **"Sales"** in the dataset, convert these two columns to numeric type. 

```python
orders['Profit'] = orders['Profit'].str.replace(',', '')
orders['Profit'] = orders['Profit'].str.replace('$', '')


orders['Sales'] = orders['Sales'].str.replace(',', '')
orders['Sales']= orders['Sales'].str.replace('$', '')


orders['Profit'] = pd.to_numeric(orders['Profit'])
orders['Sales'] = pd.to_numeric(orders['Sales'])
```

### Problem 2: Inventory Management
- Retailers that depend on seasonal shoppers have a particularly challenging job when it comes to inventory management. Your manager is making plans for next year's inventory.
- He wants you to answer the following questions:
    1. Is there any seasonal trend of inventory in the company?
    2. Is the seasonal trend the same for different categories?

- ***Hint:*** For each order, it has an attribute called `Quantity` that indicates the number of product in the order. If an order contains more than one product, there will be multiple observations of the same order.

```python
orders['Order.Date'] = pd.to_datetime(orders['Order.Date'])
orders['Month'] = orders['Order.Date'].dt.month
orders['Month']
```

```python
orders.groupby(['Month','Category']).agg({'Quantity':'sum'}).sort_values('Month')
```

```python
df1 = orders.groupby(['Month','Category']).agg({'Quantity':'sum'}).sort_values('Month').unstack()
df1.plot.bar()
```

```python
df2 = orders.groupby('Month').agg({'Quantity':'sum'}).sort_values('Month')
df2.plot.bar()
```

### Problem 3: Why did customers make returns?
- Your manager required you to give a brief report (**Plots + Interpretations**) on returned orders.

	1. How much profit did we lose due to returns each year?


	2. How many customer returned more than once? more than 5 times?


	3. Which regions are more likely to return orders?


	4. Which categories (sub-categories) of products are more likely to be returned?

- ***Hint:*** Merge the **Returns** dataframe with the **Orders** dataframe using `Order.ID`.

```python
returns
```

```python
merged = pd.merge(orders, returns, how = 'inner',left_on = 'Order.ID',right_on = 'Order ID')
```

```python
merged.columns
```

```python
#profit lost each year duer to returns


merged['Year'] = merged['Order.Date'].dt.year
Profit_lost_per_year = merged.groupby('Year').agg({'Profit':'sum'})
Profit_lost_per_year
```

```python
#customers who returned more than once


customer = merged.groupby('Customer.ID').agg({'Order.ID':'count'})
more_than_1 = customer[customer['Order.ID'] > 1].sort_values('Order.ID', ascending = False)
```

```python
#customers who returned more than five times

more_than_5 = customer[customer['Order.ID'] > 5].sort_values('Order.ID', ascending = False)
```

```python
#number of orders returned per region
returns_reg = merged.groupby('Region_x').agg({'Order.ID':'count'}).sort_values('Order.ID', ascending = False)
returns_reg

```

```python
orders_reg = orders.groupby('Region').agg({'Order.ID':'count'}).sort_values('Order.ID', ascending = False)
orders_reg

region_merged = pd.concat([orders_reg, returns_reg], axis = 1)
region_merged

region_merged['ratio'] = region_merged.iloc[:,1]/region_merged.iloc[:,0]
region_merged.sort_values('ratio', ascending = False)

```

```python
#this checks which sub category is most likely to be returned
categories_ret = merged.groupby('Sub.Category').agg({'Order.ID':'count'}).sort_values('Order.ID', ascending = False)
categories_ret

categories_ord = orders.groupby('Sub.Category').agg({'Order.ID':'count'}).sort_values('Order.ID', ascending = False)
categories_ord

category_merged = pd.concat([categories_ord, categories_ret], axis = 1)
category_merged

category_merged['ratio'] = category_merged.iloc[:,1]/category_merged.iloc[:,0]
category_merged.sort_values('ratio', ascending = False)
```
