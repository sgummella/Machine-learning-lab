```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", 500)
orders = pd.read_csv("Orders.csv")
returns = pd.read_csv("Returns.csv")
```

```python
orders.Profit = orders.Profit.str.replace("$", "")
orders.Profit = orders.Profit.str.replace(",", "")
orders.Sales = orders.Sales.str.replace("$","")
orders.Sales = orders.Sales.str.replace(",", '')
orders['Sales'] = pd.to_numeric(orders['Sales'])
orders['Profit'] = pd.to_numeric(orders['Sales'])
```

```python
all_categorical = orders.select_dtypes(exclude=[np.number])
for col in all_categorical.columns:
    print(all_categorical[str(col)].value_counts())
    print('\n')
```

```python
orders['Month'] = orders['Order.Date'].apply(lambda x: x.split('/')[0]).astype(int)
orders['Year'] = orders['Order.Date'].apply(lambda x: x.split('/')[2]).astype(int)
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (15,7))
eda1 = orders.groupby(['Month','Category']).agg({'Quantity':'sum'}).sort_values('Month').unstack().plot(ax=ax)
```

```python
plt, ax = plt.subplots(figsize = (15,7))
eda2 = orders.groupby(['Month','Sub.Category']).agg({'Quantity': 'sum'}).sort_values('Month').unstack().plot(ax=ax)
```

```python
returns.head()
```

```python
joined = orders.merge(returns, how='inner', left_on = 'Order.ID', right_on = 'Order ID')
joined.head()
```

```python
lost = joined.groupby("Year").agg({'Profit':'sum'}).rename(columns = {"Profit": "Loss"})
lost
```

```python
freq_returns = joined.groupby(["Customer.ID", 'Order.ID']).agg({"Order ID": 'count'}).sort_values("Order ID", ascending = False)
freq_returns.head()
```

```python
freq_returns_1 = freq_returns[freq_returns['Order ID'] > 1].count()
freq_returns_5 = freq_returns[freq_returns['Order ID'] > 5].count()
```

```python
region_ord = orders.groupby("Region").agg({"Order.ID": "count"}).sort_values("Order.ID", ascending = False)
region_ret = joined.groupby('Region_y').agg({"Order ID": 'count'}).sort_values("Order ID", ascending= False)
```

```python
regions = pd.concat([region_ord, region_ret], sort=True, axis=1).rename(columns = {"Order.ID": "Order", "Order ID": "Return"})
regions['Percent Returned'] = regions['Return']/regions['Order'] * 100
regions = regions.sort_values("Percent Returned", ascending = False)
regions
```

```python
subc_ord = orders.groupby("Sub.Category").agg({"Order.ID": "count"}).sort_values("Order.ID", ascending = False)
subc_ret = joined.groupby("Sub.Category").agg({"Order ID": "count"}).sort_values("Order ID", ascending = False)
```

```python
subc = pd.concat([subc_ord, subc_ret], sort = True, axis = 1).rename(columns = {"Order.ID": "Order", "Order ID": "Return"})
subc['Percent Returned'] = subc['Return']/subc['Order'] * 100
subc = subc.sort_values("Percent Returned", ascending = False)
subc
```
